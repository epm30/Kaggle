### 나라
나라마다 다른 형태의 트렌드를 갖고있으며 나이나 교육수준이 다를것으로 기대된다.

```python
mcr.columns
```

```python
# 각 나라별 응답자수를 합쳐서 데이터 프레임으로 만드는 역할을 한다.
coun_df = pd.DataFrame(mcr["Country"].value_counts())
# index 칼럼을 country라는 칼럼으로 생성한다.
coun_df['country'] = coun_df.index
# 컬럼들에 이름을 응답자수, 국가로 변경한다.
coun_df.columns = ['응답자수', '국가']
# reset_index() index칼럼을 초기화해 넘버링을 해주고 index 칼럼을 삭제한다.
coun_df = coun_df.reset_index().drop('index', axis = 1)
coun_df.head(10)
```

가장 많은 곳은 미국이고 그뒤로는 인도, 러시아, 영국, 중국..

### 자주 받는 질문
초보자에게 가장 많이 받는 질문이 무엇인지 알아보도록 하자.

본인의 경우 데이터 공부를 시작할때 맨 처음 고민했던건 프로그래밍 언어 중 R과 파이썬을 고민했었다.
따라서 그 부분을 찾아보자

```python
sns.countplot(y = 'LanguageRecommendationSelect', data = mcr)
pass
```

파이썬이 R보다 압도적으로 많이 이용되고 있다. 

```python
top_lang = mcr['LanguageRecommendationSelect'].value_counts()
top_lang_dist = []
# 언어들중 나이에 미기재된 값을 제거하고, 특정 언어를 추가한다.
for lang in top_lang.index:
    top_lang_dist.append(mcr[(mcr["Age"].notnull()) &
                             (mcr['LanguageRecommendationSelect'] == lang)]['Age'])
group_labels = top_lang.index


# 여기서부터
fig = fig_fact.create_distplot(top_lang_dist, group_labels, show_hist=False)
py.iplot(fig, filename='Language Preferences by Age')
# 내 실력으로는 역부족.. 좀더 공부해보도록하자
```

우리는 프로그래밍언어의 종류와 나이분포와 겹치는경향이 있는것을 알수있다. 즉 특별한 경향성이 없다고 할수있다.
